Implementación ecuación (79) del paper 

[1] G. Diebold and T. Sun, “PROPERTIES OF PHOTOACOUSTIC WAVES IN ONE, TWO, AND THREE DIMENSIONS,” Acustica, 1994, Accessed: Jun. 04, 2024. [Online]. Available: https://www.semanticscholar.org/paper/PROPERTIES-OF-PHOTOACOUSTIC-WAVES-IN-ONE%2C-TWO%2C-AND-Diebold-Sun/e66e8acd5e7cf6d42f94ece87876b6c9d0a447d3



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft, signal, integrate, special
import sys
```

Sea $\rho_0$ el radio del cilindro, y $c$ la velocidad del sonido. Definimos:
1) $\hat{\rho}=\frac{\rho}{\rho_0}$, coordenada espacial generalizada.
2) $\hat{t}=\frac{c}{\rho_0}t$, tiempo genereralizado
3) $\hat{q}=\frac{\omega\rho_0}{c}$, frecuencia generalizada
4) $\hat{\tau}=\frac{c}{\rho_0}\left[t-(\rho-\rho_0)/c\right]$, retardo generalizado

Se tiene que
$$ p(\hat{t})=\frac{K}{4}\int_{-\infty}^\infty J_1(\hat{q})H_0^{[1]}(\hat{q}\hat{\rho})e^{-i\hat{q}\hat{t}}d\hat{q}$$

donde $K=\frac{\alpha E_0\beta c^2}{C_p}$ con:
1) $\alpha$: Coeficiente de absorción óptica.
2) $E_0$: Fluencia ($J/m^2$).
3) $\beta$: Coeficiente de expansión isobárica.
4) $c$: Velocidad del sonido
5) $C_p$: Capacidad de calor por unidad de masa.


Podemos escribir $\hat{\tau}=\hat{t}-\hat{\rho}+1$ y:

$$ p(\hat{\tau}+\hat{\rho}-1)=\frac{K}{4}\int_{-\infty}^\infty J_1(\hat{q})H_0^{[1]}(\hat{q}\hat{\rho})e^{-i\hat{q}\hat{t}}d\hat{q}$$


Sea $B$ el máximo ancho de banda observable en Hz. Sea $W=2\pi B\rho_0/c$ el ancho de banda generalizado. Suponemos que muestreamos a una tasa de muestreo generalizado $\hat{T}=T \frac{c}{\rho_0}$ exactamente igual al máximo permitido por Nyquist $\hat{T}=\frac{\pi}{W}=\frac{c}{\rho_0}\frac{1}{2B}$ 

```python
#Seteo valores para constante K y y también c que tiene una influencia un poco más relevante porque con  ella se define el retardo generalizado. Lo mismo para el radio del cilindro
#rho_0 y para el ancho de banda B.
c=1500
E0=1
beta=1
alpha=1
Cp=1
rho_0=1e-3
K=alpha*E0*beta*c*c/Cp
B=20e6
W=2*np.pi*B*rho_0/c
T=1/2/B

#Punto campo (posición del sensor en donde se calculara la onda de presión)
rho=20e-3
rhohat=rho/rho_0
tmax= 1*(rho+rho_0)/c
```

Pongo todo en una función con comentarios

```python
def cylinder_sinogram_Diebold(c,E0,beta,alpha,Cp,rho_0,rho,T,tmax):
    #Esta función implementa la ecuación (79) del paper
    #G. Diebold and T. Sun, “PROPERTIES OF PHOTOACOUSTIC WAVES IN ONE, TWO, AND THREE DIMENSIONS,” Acustica, 1994, Accessed: Jun. 04, 2024. [Online]. 
    # Available: https://www.semanticscholar.org/paper/PROPERTIES-OF-PHOTOACOUSTIC-WAVES-IN-ONE%2C-TWO%2C-AND-Diebold-Sun/e66e8acd5e7cf6d42f94ece87876b6c9d0a447d3
        
    
    #PARAMETROS
    
    # c= Velocidad del sonido
    
    # E0= Fluencia
    
    # beta= Coeficiente de expansión isobárica
    
    # alpha= Coeficiente de absorpción óptica
    
    # Cp= Capacidad de calor por unidad de masa
    
    # rho_0= Radio del cilindro iluminado por láser. Se supone que el mismo está en origen
    
    # rho= Punto campo donde se medirá la presión con respecto al origen (donde está el cilindro de muestra excitado). Este párametro es simplemente la distancia radial
    #      desde el origen a la posición del sensor. No es necesario especificar ángulo porque la radiación emitida es simétrica con respecto a phi. En el caso de llamar la función
    #      para distintos puntos campos (sensores) se deberá externamente computar la distancia radial desde el centro del cilindro emisor a la posición deseada, ya que los detalles
    #      dependen de la geometría del arreglo de sensores.
    
    # B = Ancho de banda en Hz del sistema de medición. 
    
    # T= Período de muestreo del sistema de adquisición. Típicamente T< 1/2B, donde B es el ancho de banda de los sensores en Hz
    
    #tmax= Tiempo máximo de la onda de presión a calcular. Este es un parámetro que se debe calcular externamente ya que depende de la geometria de la muestra y del arreglo de sensores. En teoría cuando 
    #      el cilindro fuente es infinito la presión en cualquier punto de sensado no se hace cero nunca. Este valor habria que elegir con criterio. Por ejemplo podemos considerar la distancia máxima entre puntos de la muestra y puntos
    #      de sensado, dividir dicho número por la velocidad del sonido y multiplicarla por un margen de seguridad (ej: 1.2). Por todas estás razones es que es mejor pasar este número como un parámetro
    
    ####TO DO####
    # 1)Incluir efecto de filtrado del sistema de adquisición mediante la especificación de la correspondiente respuesta en frecuencia!
    
    #Definimos algunos parámetros
    K=alpha*E0*beta*c*c/Cp       #Constante multiplicativa para la onda de presión
    W=2*np.pi*rho_0/c/2/T        #Ancho de banda generalizado en radianes. Es necesario para el correcto calculo de la FFT
    rhohat=rho/rho_0             #Punto campo normalizado con respecto al radio del cilindro emisor.
    

    power=np.ceil(np.log2(tmax/T))   #Potencia de 2 más cercana al número total de muestras a usar para representar la onda presión en el punto campo
    N=int(2**power)                  #Número de muestras para la FFT y para la representación temporal de la onda de presión
    qhat = np.linspace(-W, W,N,endpoint=False)  
    D=round(N/2)                     #Punto medio del muestreo en la frecuencia. Corresponde idealmente a frecuencia 0 Hz. Se necesita para fijar valor en dicha frecuencia de la señal 
                                     #a invertir ya que hay una singularidad evitable en dicho punto
    tmax_mod=N*T                     #Como elijo el menor N que sea potencia de dos (pero mayor a tmax/T), debo modificar el tmax, que siempre va a ser mayor igual al pasado como parámetro a la función    
    
    J1=special.j1(qhat)
    H0=special.hankel1(0,rhohat*qhat)
    N1=K/4*J1*H0                         #Función a invertir
    N1[D]=0                              #Fijo valor en 0 Hz
        
    #Inversion via FFT
    
    N2=fft.fftshift(N1)
    y=fft.fft(N2)                        #Señal temporal en el punto campo
    
    
    #Esto se puede sacar en el futuro para acelerar el calculo. Ahora lo dejo solo para despues poder graficar fácilmente los resultados obtenidos
    
    t= np.linspace(0,tmax_mod,N,endpoint=False) #Instantes de muestreo verdaderos de la señal de presión 
    tau= c*t/rho_0- rhohat+1                #Tiempos normalizados tal como se definen en el paper de Diebold. Util para graficar y comparar con los resultados del paper, pero nada más.
    
    return tau, t, np.real(y)
```

```python
import timeit
n=10
result = timeit.timeit(stmt='cylinder_sinogram_Diebold(c,E0,beta,alpha,Cp,rho_0,rho,T,tmax)', globals=globals(), number=n)
print(f"Execution time is {result / n} seconds")
```

```python
tn,tt,P = cylinder_sinogram_Diebold(c,E0,beta,alpha,Cp,rho_0,rho,T,tmax)
```

```python
plt.figure(figsize=(15, 10))
plt.xlabel('Normalized Time')
plt.ylabel('Amplitude')
plt.plot(tn, P)
plt.show()
```
