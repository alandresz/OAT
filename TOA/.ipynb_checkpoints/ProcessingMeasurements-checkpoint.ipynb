## Measurement: venas 5x5 mm2 sample: med12oct23_5 (T = 25 us) 180 angles

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from OAT import applyDAS, applyUBP, createForwMatdotdet, createForwMatdotdetMIR, createForwMatdotdetLBW
from scipy.interpolate import interp1d
```

```python
filename = 'Mediciones/Mediciones 12-10-2023/med12oct23_6.npz'

data = np.load(filename)
t=data['t']; sinogram=data['sinogram']; angles=data['angles']; Rto=data['Rto']; arco=data['arco']
Rs=data['Rs']; Fl=data['Fl']; Navg=data['Navg']; sample=data['sample']; vs=data['vs']

Ns, Nt = sinogram.shape
to = t[0]
tf = t[-1]
nx = 128
dx = 100e-6
p = -1*sinogram
dsa = Rs
fmax = 4e6
```

```python
Xdas = applyDAS(Ns,Nt,dx,nx,dsa,arco,vs,to,tf,p*1)
axis_ticks = [-nx*dx*1e3/2, nx*dx*1e3/2, -nx*dx*1e3/2, nx*dx*1e3/2]
plt.figure(),plt.imshow(np.reshape(Xdas,(nx,nx)),extent=axis_ticks, aspect="auto")
plt.xlabel("x [mm]")
plt.ylabel("y [mm]")
plt.title(sample)
plt.colorbar()
```

```python
Ntn = 512
#Nsn = 36
#Ao0 = createForwMatdotdet(Ns,Ntn,dx,nx,dsa,arco,vs,to,tf)
#Ao1 = createForwMatdotdetLBW(Ns,Ntn,dx,nx,dsa,arco,vs,to,tf,True)
Ao2 = createForwMatdotdetMIR(Ns,Ntn,dx,nx,dsa,arco,vs,to,tf)
```

```python
tn = np.linspace(to,tf,Ntn)
pn = np.zeros((Ns,Ntn));
for i1 in range(0,Ns):
    fp=interp1d(t,p[i1,:]);
    aux=fp(tn);
    pn[i1,:]=aux;
```

```python
Ao = Ao2
Xlbp = Ao.T@(-1*pn.ravel())
axis_ticks = [-nx*dx*1e3/2, nx*dx*1e3/2, -nx*dx*1e3/2, nx*dx*1e3/2]
plt.figure(),plt.imshow(np.reshape(Xlbp,(nx,nx)),extent=axis_ticks, aspect="auto")
plt.xlabel("x [mm]")
plt.ylabel("y [mm]")
plt.title(sample)
plt.colorbar()
```

```python
np.save('venas5x5.npy',Xlbp)
```
